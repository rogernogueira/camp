```python
import pandas as pd
```

```python
df = pd.read_excel('/pessoas.xlsx')
```

```python
def one_zero(sex):
  if sex == 'M':
    return 1
  else:
    return 0
```

```python
def calc_imc(peso, altura):
  imc = peso / (altura * altura)
  return imc
```

```python
df["sexo_binario"] = df["sexo"].apply(one_zero)
```

```python
df['imc'] = df.apply(lambda x : calc_imc(x['peso'],x['altura']), axis=1)
```
